**Dictionaries**

> *treasure (noun): a collection of valuable objects such as gold, silver, jewels, or other precious items.*

A dictionary in Python is a built-in data structure that stores data in key-value pairs.

Dictionaries are written with curly brackets, and have key-value pairs are separated by comma.

**Key Features of Dictionaries**

- ***Unordered vs Ordered:*** Items in a dictionary are not stored in a specific order (before Python 3.7). Since Python 3.7+, dictionaries maintain insertion order.

- ***Mutable:*** You can modify, add, or remove key-value pairs after creation.

- ***Key-Value Pairs:*** Each key is associated with a value, making data retrieval efficient.

- ***Unique Keys:*** No duplicate keys are allowed; each key in a dictionary must be unique.

```python
student = {
    "name": "Izzah",
    "major": "Computer Science",
    "semester": "6th"
}
print(student)
```

```python
empty_dict = {}
print(empty_dict)
```

**Keys in a Dictionary**
- ***Definition:*** Keys are unique identifiers used to access values in a dictionary.
- ***Uniqueness:*** No two keys can be the same in one dictionary. Key is a unique identifier that is used to map a specific value present against it in the dictionary.
- ***Immutable:*** Keys must be of an immutable data type.

 Common valid key types include:  
  - Strings ("name")
  - Numbers (integers, floats)
  - Tuples (if all elements inside are also immutable)
- ***Case-Sensitive:*** "name" and "Name" are treated as different keys.
- ***Order (Python 3.7+):*** Dictionaries preserve the order in which keys are added.

```python
my_dict = {
    "name": "string",
    1: "integer",
    (2, 3): (1,2,3),
    1: "float"
}
print(my_dict)

# a = "ahmad"
# a = "ali"
# print(a)
```

**Values in a Dictionary**
- ***Definition:*** Values are the data associated with each key.
- ***Value Data Type:*** Values can be of any datatype:
  - Strings
  - Numbers
  - Lists
  - Tuples
  - Other Dictionaries (Nested dictionaries)
  - Even functions or objects
- ***Duplicate Values Allowed:*** Multiple keys can have the same value, even though keys must be unique.
- ***Mutable:*** You can change a value using its key.

```python
s = {
    "name": "Noor",
    "semester": 6,
    "major": "Computer Science",
    "university": "UAF",
    "institution": "UAF"
}
print(s, "\n")

s["semester"] = 4

print(s)
```

**Length of a Dictionary**
- The length of a dictionary refers to how many key-value pairs it contains.
- You can find the length using the `len()` function.

```python
student1 = { 1 : "Tayyaba",
         "age" : 21,
         "Country": "Pakistan",
         "city" : "Faisalabad",
          "age":23,
          "age":25
         }
print(student1)
print("\n", len(student1))
```

**Accessing Elements in a Dictionary**

In Python, you can access the values stored in a dictionary using their keys.

**Ways to Access Elements**
- Using Square Brackets `[]` (Direct Access)
- Using `.get()` Method (Safer Access)
- Accessing All Keys, Values, or Both

**1- Using Square Brackets `[]`**
- The most common way to access a value.
- You provide the key inside square brackets.
- If the key does not exist, Python raises a KeyError.
- Can't use Slicing

```python
car = {
    "brand": "Toyota",
    "model": "Corolla",
    "year": 2020
}

print(car["brand"])
print(car["model"])
print(car["color"])
```

**2- Using `.get()` Method**
- Works like square brackets, but doesn't raise an error if the key is missing.
- Returns None if the key does not exist (you can also set a custom default value).

```python
book = {
    "title": "Data Science Essentials",
    "author": "Alice Brown"
}

print(book.get("title"))
print(book.get("publisher"))
print(book.get("publisher", "David"))
```

**3- Accessing All Keys, Values, or Both**



```python
book = {
    "title": "Data Science Essentials",
    "author": "Alice Brown"
}

print(book.keys())
print(book.values())
print(book.items()) #get all items as tuples
```

**Adding and Changing Elements of a Dictionary**

In Python, you can add new key-value pairs to a dictionary easily by directly assigning a value to a new key.



**1- Direct Assignment (Most Common Way)**
- If the key does not exist, a new key-value pair is added.
- If the key already exists, the value is updated.

```python
stu = {
    "name": "Rafia",
    "age": 21
}

stu["major"] = "Computer Science"  #adding
stu["age"] = 25   #changing
print(stu)
```

**2- Using the `update()` Method**
- You can add multiple key-value pairs at once using `update()`.
- If the key already exists, its value will be updated.

```python
stu = {
    "name": "Ali",
    "age": 21,
    "major": "Computer Science"
}
print(stu)

stu.update( {"university": "UAF",
             "major": "Software Engineering",
             "city": "Faisalabad"} )
print(stu)
```

**Removing Elements from a Dictionary**

You can remove key-value pairs from a dictionary using several methods.

**1- Using** `pop()` **Method**
- Removes a specific key and returns its value.
- Raises KeyError if the key does not exist (unless a default value is provided).

```python
trip = {
    "destination": "Turkey",
    "duration": "10 days",
    "budget": 1500,
    "travel_mode": "Flight"
}

# trip.pop("budget")
# print(trip)

# trip.pop("city")
trip.pop("city", "Not Found")
```

**2- Using `del` Statement**
- Deletes a specific key-value pair.
- Raises KeyError if the key does not exist.

```python
student = {"name": "Izzah",
           "age": 21,
           "major": "CS"}

del student["major"]

print(student)

del student
print (student)
```

**3- Using `popitem()` Method**
- Removes and returns the last inserted key-value pair.
- It takes no argument.
- Works well with Python 3.7+ where dictionaries maintain insertion order.
- Raises KeyError if the dictionary is empty.

```python
food_order = {"dish": "Pizza",
              "quantity": 2,
              "price": 3000}

print("Removed Item:")
food_order.popitem()
```

**4- Using `clear()` Method**
- Removes all key-value pairs, leaving the dictionary empty.

```python
food = {"fruit": "Mango",
        "vegetable": "Potato",
        "meat": "Mutton"}

food.clear()
print(food)
```

**Membership Operators in Dictionaries**
- Membership operators only check for keys, not values.
- If you want to check for a value, you need to do it differently, using the `values()` method.

```python
student2 = {
    "name": "Izzah",
    "age": 21,
    "major": "Computer Science"
}

print("name" in student2)

print("university" in student2)

print("age" not in student2)

print("Computer Science" in student2.values())
```

**Iterating Through a Dictionary**


**1- Iterating Through Keys (Default Way)**
- When you directly iterate through a dictionary, you are iterating over its keys.

```python
book = {"title": "Python Programming",
        "author": "John Doe",
        "year": 2022}

for key in book:
    print(key)
```

```python
record = {"name" : "Tayyaba",
         "age" : 21,
         "country" : "Pakistan",
         "city" : "Faisalabad",
        }

for i in record:
  print(f"{i}: {record[i]}")
```

**2- Iterating Through Keys (Explicitly)**
- Using `.keys()` method.
- This is the same as the above, but more explicit.

```python
data = {
    "destination": "Maldives",
    "departure_date": "2025-06-15",
    "return_date": "2025-06-25",
    "airline": "Emirates"
}

for key in data.keys():
    print(key)
```

**3- Iterating Through Values**
- Using `.values()`
- To loop over values only.

```python
languages = {
    1 : "Python",
    2 : "C++",
    3 : "Java",
    4: "Ruby",
    5 : "C",
}

for value in languages.values():
  print(value)
```

**4- Iterating Through Key-Value Pairs**
- Using `.items()`
- To loop over both keys and values together.

```python
languages = {
    'one' : "Python",
    'two' : "C++",
    'three' : "Java",
    'four': "Ruby",
    'five' : "C#",
}

for key, value in languages.items():
  print(f"{key}: {value}")
```

**Nested Dictionary**
- A nested dictionary is a dictionary inside another dictionary.
- This allows you to store more complex data.

```python
students = {
    "student1": {"name": "Ali", "age": 23, "major": "CS"},
    "student2": {"name": "Ahmad", "age": 22, "major": "IT"}
}

# Accessing a nested value
print(students["student1"]["age"])
```

**Use of `zip()` with `dict()`**

The `zip()` function can combine two lists (or other iterables) into pairs of key-value pairs, which can then be directly converted into a dictionary using `dict()`.

```python
keys = ['name', 'age', 'city']
values = ['Alice', 25, 'New York']

result = dict(zip(keys, values))

print(result)
```

```python
keys = ['name', 'age', 'city']
values = ['Alice', 25]

result = dict(zip(keys, values))

print(result)
```

```python
keys = ('name', 'age', 'city')
values = ('Alice', 25, 'New York')

result = dict(zip(keys, values))

print(result)
```

```python
x = dict(zip((1,2,3),(5,6,7)))
print(x)
```

```python
keys = {'name', 'age', 'city'}
values = {'Alice', 'New York', 25}

result = dict(zip(keys, values))

print(result)
```
